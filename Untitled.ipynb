```python
features = df.drop('output',axis=1)
target = df['output']
x_train,x_test,y_train,y_test= train_test_split(features,target,test_size=0.2,random_state=123)
```

```python
scaller = MinMaxScaler(feature_range=(0,1))
xtr_scalled = pd.DataFrame(data=scaller.fit_transform(x_train),columns=x_train.columns)
xts_scalled = pd.DataFrame(data=scaller.transform(x_test),columns=x_test.columns)
```

```python
lg = LogisticRegression()
lg.fit(xtr_scalled,y_train)
prediction = lg.predict(xts_scalled)
score = f1_score(y_test,prediction)
accuracy = accuracy_score(y_test,prediction)
precision = precision_score(y_test,prediction)
recall = recall_score(y_test,prediction)
m1=pd.DataFrame({'f1score':score,'recall':recall,'precision':precision,'accuracy':accuracy},index=['LogisticRegression'])
m1
```

```python
tree = DecisionTreeClassifier()
tree.fit(xtr_scalled,y_train)
prediction = tree.predict(xts_scalled)
score = f1_score(y_test,prediction)
accuracy = accuracy_score(y_test,prediction)
precision = precision_score(y_test,prediction)
recall = recall_score(y_test,prediction)
m2=pd.DataFrame({'f1score':score,'recall':recall,'precision':precision,'accuracy':accuracy},index=['DecisionTree'])
m2
```

```python
fr = RandomForestClassifier()
fr.fit(xtr_scalled,y_train)
prediction = fr.predict(xts_scalled)
score = f1_score(y_test,prediction)
accuracy = accuracy_score(y_test,prediction)
precision = precision_score(y_test,prediction)
recall = recall_score(y_test,prediction)
m3=pd.DataFrame({'f1score':score,'recall':recall,'precision':precision,'accuracy':accuracy},index=['RandomFOrest'])
m3
```

```python
knn = KNeighborsClassifier()
knn.fit(xtr_scalled,y_train)
prediction = knn.predict(xts_scalled)
score = f1_score(y_test,prediction)
accuracy = accuracy_score(y_test,prediction)
precision = precision_score(y_test,prediction)
recall = recall_score(y_test,prediction)
m4=pd.DataFrame({'f1score':score,'recall':recall,'precision':precision,'accuracy':accuracy},index=['KNeighbors'])
m4
```

```python
XGB = XGBClassifier()
XGB.fit(xtr_scalled,y_train)
prediction = XGB.predict(xts_scalled)
score = f1_score(y_test,prediction)
accuracy = accuracy_score(y_test,prediction)
precision = precision_score(y_test,prediction)
recall = recall_score(y_test,prediction)
m5=pd.DataFrame({'f1score':score,'recall':recall,'precision':precision,'accuracy':accuracy},index=['XGB'])
m5
```

```python
result_model_1 = pd.concat([m1,m2,m3,m4,m5],axis=0)
result_model_1
```
